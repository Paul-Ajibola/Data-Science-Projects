```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Creating the experimental dataset
class0 = np.random.randn(50, 2) + np.array([-2, -2])

class1 = np.random.randn(50, 2) + np.array([2, 2])

X = np.vstack((class0, class1))
y = np.array([0]*50 + [1]*50)

print('X_shape:', X.shape)
y.shape
```

```python
#  Initialize weights and bias
w = np.zeros(X.shape[1])
b = 4.0

# Set up learning rate and epochs
lr = 0.1
epochs = 5
errors_per_epoch = []

# Step function
def step(z):
    return 1 if z >= 0 else 0


# Training loop
for epoch in range(epochs):
    total_errors = 0
    for xi, yi in zip(X, y):
        z = np.dot(w, xi) + b
        y_hat = step(z)

        # Percetron update rule
        error = yi - y_hat
        if error != 0:
            w += lr * error * xi
            b += lr * error
            total_errors += 1

    errors_per_epoch.append(total_errors)
    print(f"Epoch {epoch+1}: errors = {total_errors}")
```

Plot training errors over epochs

```python
plt.plot(range(1, epochs+1), errors_per_epoch, marker='o')
plt.title('Number of Misclassification over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Number of Errors')
plt.grid(True)
plt.show()
```

The downward trend of the line shows that the model is converging. With each drop in the line, it is clear that the perceptron successfully updates it weights and bias to correctly classify more points than it did in the previous pass. The perceptron begins with 10 errors in epoch 1, and the error count drops sharply to 2 in epoch 2 and down to 1 in epoch 3. In epoch 4, the line hits the x-axis, meaning the number of misclassifications reached 0. It maintained 0 errors through epoch 5, confirming the model has finifhed its training successfully. 
Therefore, the perceptron is a good and has found a perfect linear boundary. Also, the two classes in the dataset are linearly separable.
